We’ll implement a **minimal but faithful SDNE** following the paper:

* adjacency rows as input
* deep autoencoder
* second-order loss (reconstruction with β)
* first-order loss (local structure via Laplacian)
* L2 regularization

```python
import torch
import torch.nn as nn
import torch.optim as optim
import networkx as nx
```

---

## Step 1 – Build a Graph with NetworkX

For a demo, let’s just use the classic karate club graph:

```python
# Example graph (undirected, unweighted)
G = nx.karate_club_graph()
n_nodes = G.number_of_nodes()
print("Number of nodes:", n_nodes)
```

---

## Step 2 – Build the Adjacency Matrix (S)

From the paper, SDNE takes the **adjacency row** of each node as input. 
So we build a dense adjacency matrix $S \in \mathbb{R}^{n \times n}$.

```python
# Create adjacency matrix S (n x n), as float tensor
# Nodes are assumed labelled 0..n-1
A = nx.to_numpy_array(G, nodelist=sorted(G.nodes()))  # shape (n, n)
S = torch.tensor(A, dtype=torch.float32)
```

## Step 3 – Precompute Laplacian L and Weight Matrix B

### 3.1 First-order proximity → Laplacian (L = D - S)

This will be used for the **first-order loss** (local structure)

```python
# Degree matrix D
degrees = S.sum(dim=1)          # (n,)
D = torch.diag(degrees)         # (n, n)

# Graph Laplacian L = D - S
L = D - S                       # (n, n)
```

### 3.2 Weight matrix (B) for second-order loss

The second-order loss reconstructs adjacency rows and **weights non-zero entries by β > 1**. 

```python
beta = 5.0  # can tune this

B = torch.ones_like(S)
B[S > 0] = beta    # non-zero entries get larger penalty
```

---

## Step 4 – Define the SDNE Model (Deep Autoencoder)

We follow the paper’s idea: **multiple nonlinear layers** in encoder + mirrored decoder. 

For simplicity, we’ll use a **2-hidden-layer encoder**:

* Input: dimension = number of nodes (n)
* Hidden1: e.g. 100
* Embedding: e.g. 16
* Decoder: mirror the encoder

```python
class SDNE(nn.Module):
    def __init__(self, n_nodes, hidden_layers=[100, 16]):
        super(SDNE, self).__init__()
        # hidden_layers: [h1, h2,..., emb_dim]; we’ll use last as embedding

        dims = [n_nodes] + hidden_layers  # encoder dims
        # Encoder
        enc_layers = []
        for i in range(len(dims) - 1):
            enc_layers.append(nn.Linear(dims[i], dims[i+1]))
            enc_layers.append(nn.Sigmoid())  # as in the paper
        self.encoder = nn.Sequential(*enc_layers)

        # Decoder: mirror of encoder (reverse dims)
        rev_dims = list(reversed(dims))
        dec_layers = []
        for i in range(len(rev_dims) - 1):
            dec_layers.append(nn.Linear(rev_dims[i], rev_dims[i+1]))
            # last layer: no activation (we’ll reconstruct real-valued adjacency)
            if i < len(rev_dims) - 2:
                dec_layers.append(nn.Sigmoid())
        self.decoder = nn.Sequential(*dec_layers)

    def forward(self, x):
        # x: (n, n)
        y = self.encoder(x)      # embeddings
        x_hat = self.decoder(y)  # reconstruction
        return y, x_hat

    def get_embeddings(self, x):
        with torch.no_grad():
            y = self.encoder(x)
        return y
```

---

## Step 5 – Define SDNE Loss Functions

Total loss:

$$
L_{\text{mix}} = L_{2nd} + \alpha L_{1st} + \nu L_{reg}
$$

### 5.1 Second-order loss (L_{2nd}) (adjacency reconstruction)

$$
L_{2nd} = |(\hat{X} - X) \odot B|_F^2
$$


```python
def loss_second_order(X, X_hat, B):
    # X, X_hat, B: (n, n)
    diff = (X_hat - X) * B
    return torch.sum(diff * diff)  # Frobenius norm squared
```

### 5.2 First-order loss $L_{1st}$ (local structure)

From the paper:

$$
L_{1st} = \sum_{i,j} s_{i,j} |y_i - y_j|^2 = 2 , \text{tr}(Y^T L Y)
$$ 

We’ll implement the Laplacian form:

```python
def loss_first_order(Y, L):
    # Y: (n, d), L: (n, n)
    # 2 * trace(Y^T L Y)
    LY = torch.matmul(L, Y)          # (n, d)
    YtLY = torch.sum(Y * LY)         # trace of Y^T L Y
    return 2 * YtLY
```

### 5.3 Regularization (L_{reg})

Sum of squared weights for all linear layers:

```python
def loss_reg(model):
    reg = 0.0
    for param in model.parameters():
        reg = reg + torch.sum(param * param)
    return reg
```

### 5.4 Combine them

We’ll use typical hyperparameters: α ~ 1e-3 to 1e-2, ν small.

```python
alpha = 1e-4   # balance first-order
nu = 1e-5      # regularization strength

model = SDNE(n_nodes=n_nodes, hidden_layers=[100, 16])
optimizer = optim.Adam(model.parameters(), lr=1e-3)
```

---

## Step 6 – Training Loop

We treat **all nodes in one batch** (because input is the full adjacency matrix).

```python
X = S  # adjacency as input

n_epochs = 200

for epoch in range(n_epochs):
    model.train()
    optimizer.zero_grad()

    Y, X_hat = model(X)

    # compute losses
    l_2nd = loss_second_order(X, X_hat, B)
    l_1st = loss_first_order(Y, L)
    l_reg = loss_reg(model)

    loss = l_2nd + alpha * l_1st + nu * l_reg

    loss.backward()
    optimizer.step()

    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch+1}/{n_epochs} | "
              f"L2nd: {l_2nd.item():.4f} | "
              f"L1st: {l_1st.item():.4f} | "
              f"Lreg: {l_reg.item():.4f} | "
              f"Total: {loss.item():.4f}")
```

- This corresponds to Algorithm 1 in the paper (without DBN pretraining, for simplicity).

---

## Step 7 – Extract Embeddings

Once trained, you get node embeddings from the encoder:

```python
model.eval()
embeddings = model.get_embeddings(X)  # shape: (n_nodes, embedding_dim)
print(embeddings.shape)
```

You can now use `embeddings` for:

* node classification
* clustering
* link prediction (e.g. cosine similarity)
* t-SNE / PCA visualization

# Tiny Example: Link Score from Embeddings

For two nodes i, j, you can define a simple similarity score like cosine:

```python
import torch.nn.functional as F

def link_score(embeddings, i, j):
    vi = embeddings[i]
    vj = embeddings[j]
    return F.cosine_similarity(vi.unsqueeze(0), vj.unsqueeze(0)).item()

print("Score(0, 1):", link_score(embeddings, 0, 1))
print("Score(0, 10):", link_score(embeddings, 0, 10))
# Higher score → more likely there should be an edge.
```

---

## Recap of the Pipeline

1. **Graph → adjacency matrix S** (NetworkX)
2. **Compute L (Laplacian) and B (β-weighting)**
3. Define **SDNE autoencoder** in PyTorch
4. Implement:

   * **second-order loss** (reconstruction with β)
   * **first-order loss** (using Laplacian)
   * **L2 regularization**
5. Combine into total loss and **train with Adam**
6. Extract **embeddings** from encoder and use them for tasks

# Visualization

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
```

```python
# Assuming G is your networkx.karate_club_graph()
# and nodes are 0..n-1

labels = []
for node in sorted(G.nodes()):
    club = G.nodes[node]['club']  # 'Mr. Hi' or 'Officer'
    labels.append(0 if club == 'Mr. Hi' else 1)

labels = np.array(labels)
```

```python
model.eval()
embeddings = model.get_embeddings(X)  # X = adjacency matrix S
# embeddings: torch.Tensor of shape (n_nodes, emb_dim)
```

```python
# Convert to numpy
emb_np = embeddings.detach().cpu().numpy()  # (n_nodes, emb_dim)

# Run t-SNE to 2D
tsne = TSNE(n_components=2, perplexity=10, learning_rate=200,random_state=42)
emb_2d = tsne.fit_transform(emb_np)  # (n_nodes, 2)
```

```python
plt.figure(figsize=(8, 6))

# Two colors for the two karate clubs
for lbl, color, name in [(0, 'tab:blue', 'Mr. Hi'), (1, 'tab:orange', 'Officer')]:
    idx = labels == lbl
    plt.scatter(
        emb_2d[idx, 0],
        emb_2d[idx, 1],
        label=name,
        alpha=0.8,
        edgecolors='k'
    )

for i, (x, y) in enumerate(emb_2d):
    plt.text(x + 0.02, y + 0.02, str(i), fontsize=8, alpha=0.7)  # optional: show node IDs

plt.title("SDNE Embeddings Visualized with t-SNE (Karate Club)")
plt.legend()
plt.xlabel("t-SNE dim 1")
plt.ylabel("t-SNE dim 2")
plt.tight_layout()
plt.show()
```
