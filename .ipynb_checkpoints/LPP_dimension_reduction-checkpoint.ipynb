```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import PCA
```

```python
class LocalityPreservingProjections:
    """
    Locality Preserving Projections (LPP) implemented in pure NumPy.

    Parameters
    ----------
    n_components : int
        Target dimensionality.
    n_neighbors : int
        Number of neighbors for kNN graph.
    use_heat_kernel : bool
        If True, use heat kernel weights; else use binary weights.
    t : float
        Heat kernel parameter (only used if use_heat_kernel=True).
    reg : float
        Regularization added to X^T D X for numerical stability.
    """

    def __init__(
        self,
        n_components=2,
        n_neighbors=10,
        use_heat_kernel=True,
        t=1.0,
        reg=1e-3,
    ):
        self.n_components = n_components
        self.n_neighbors = n_neighbors
        self.use_heat_kernel = use_heat_kernel
        self.t = t
        self.reg = reg

        self.components_ = None  # shape: (n_features, n_components)
        self.eigenvalues_ = None

    # ------------------------------------------------------------------
    # Public API
    # ------------------------------------------------------------------
    def fit(self, X):
        """
        Fit LPP on data X.

        Parameters
        ----------
        X : ndarray of shape (n_samples, n_features)
            Input data.

        Returns
        -------
        self
        """
        X = np.asarray(X, dtype=float)
        n_samples, n_features = X.shape

        # 1) Build adjacency and weight matrix W
        W = self._build_weight_matrix(X)

        # 2) Degree matrix D and graph Laplacian L
        D = np.diag(W.sum(axis=1))
        L = D - W

        # 3) Compute XLX^T and XDX^T, but with our convention X: (n_samples, n_features)
        #    Paper uses X: (n_features, n_samples), so:
        #    XLX^T (paper) = X^T L X (here)
        X_T = X.T  # shape: (n_features, n_samples)
        XLXT = X_T @ L @ X      # shape: (n_features, n_features)
        XDXT = X_T @ D @ X      # shape: (n_features, n_features)

        # 4) Regularize XDXT for stability
        XDXT_reg = XDXT + self.reg * np.eye(n_features)

        # 5) Solve generalized eigenproblem:
        #    XLXT a = λ XDXT a
        #    via whitening: let B = XDXT_reg
        #    B = Q Λ Q^T  => B^{-1/2} = Q Λ^{-1/2} Q^T
        #    Then solve eigenproblem for M = B^{-1/2}^T A B^{-1/2}
        #    where A = XLXT.
        evals_B, evecs_B = np.linalg.eigh(XDXT_reg)

        # Filter tiny eigenvalues
        tol = 1e-10
        mask = evals_B > tol
        if mask.sum() < self.n_components:
            # If too few, relax the threshold
            mask = evals_B > 0

        evals_B = evals_B[mask]
        evecs_B = evecs_B[:, mask]

        # B^{-1/2} = Q * Λ^{-1/2}
        B_inv_sqrt = evecs_B @ np.diag(evals_B ** -0.5)

        # M is symmetric
        M = B_inv_sqrt.T @ XLXT @ B_inv_sqrt
        evals, evecs = np.linalg.eigh(M)

        # Sort eigenvalues ascending (smallest have best locality preservation)
        idx = np.argsort(evals)
        evals = evals[idx]
        evecs = evecs[:, idx]

        # Take the first n_components
        n_components = min(self.n_components, evecs.shape[1])
        evecs = evecs[:, :n_components]
        self.eigenvalues_ = evals[:n_components]

        # Map back to original space:
        # a = B^{-1/2} * u   where u is eigenvector of M
        A = B_inv_sqrt @ evecs  # shape: (n_features, n_components)
        self.components_ = A

        return self

    def transform(self, X):
        """
        Project data into the LPP subspace.

        Parameters
        ----------
        X : ndarray of shape (n_samples, n_features)

        Returns
        -------
        Y : ndarray of shape (n_samples, n_components)
        """
        if self.components_ is None:
            raise RuntimeError("You must call fit(X) before transform(X).")

        X = np.asarray(X, dtype=float)
        return X @ self.components_

    def fit_transform(self, X):
        """
        Fit LPP and return the embedding of X.

        Parameters
        ----------
        X : ndarray of shape (n_samples, n_features)

        Returns
        -------
        Y : ndarray of shape (n_samples, n_components)
        """
        self.fit(X)
        return self.transform(X)

    # ------------------------------------------------------------------
    # Internal helpers
    # ------------------------------------------------------------------
    def _build_weight_matrix(self, X):
        """
        Build kNN graph weight matrix W.

        Parameters
        ----------
        X : ndarray of shape (n_samples, n_features)

        Returns
        -------
        W : ndarray of shape (n_samples, n_samples)
        """
        X = np.asarray(X, dtype=float)
        n_samples = X.shape[0]

        # Pairwise squared Euclidean distances
        # (could be memory-heavy for very large n_samples)
        # dist_ij^2 = ||x_i||^2 + ||x_j||^2 - 2 x_i·x_j
        norms = np.sum(X ** 2, axis=1, keepdims=True)  # (n_samples, 1)
        dist_sq = norms + norms.T - 2 * (X @ X.T)

        # For numerical stability, clip negative due to rounding
        dist_sq = np.maximum(dist_sq, 0.0)

        # For each point, find its k nearest neighbors (excluding self)
        k = min(self.n_neighbors + 1, n_samples)  # +1 because self is included
        # argsort along row: distances to all others
        nn_indices = np.argpartition(dist_sq, kth=k - 1, axis=1)[:, :k]

        # Build symmetric W
        W = np.zeros((n_samples, n_samples), dtype=float)

        for i in range(n_samples):
            for j in nn_indices[i]:
                if i == j:
                    continue
                d2 = dist_sq[i, j]
                if self.use_heat_kernel:
                    w_ij = np.exp(-d2 / self.t)
                else:
                    w_ij = 1.0
                # Undirected graph => symmetric W
                W[i, j] = max(W[i, j], w_ij)
                W[j, i] = max(W[j, i], w_ij)

        return W
```

```python
# 1. Generate Swiss roll data
n_samples = 1500
X, t = make_swiss_roll(n_samples=n_samples, noise=0.05)
# X: (n_samples, 3), t: "time" / unrolled coordinate for coloring
```

```python
# 2. Fit LPP to reduce from 3D -> 2D
lpp = LocalityPreservingProjections(
    n_components=2,
    n_neighbors=10,
    use_heat_kernel=True,
    t=1.0,
    reg=1e-3,
)

X_embedded = lpp.fit_transform(X)  # shape: (n_samples, 2)

print("Original shape:", X.shape)
print("Embedded shape:", X_embedded.shape)
print("LPP eigenvalues:", lpp.eigenvalues_)
```

```python
# 3. Plot original 3D Swiss roll
fig = plt.figure(figsize=(12, 5))

ax = fig.add_subplot(1, 2, 1, projection="3d")
scatter3d = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, s=5)
ax.set_title("Original Swiss Roll (3D)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# 4. Plot 2D LPP embedding
ax2 = fig.add_subplot(1, 2, 2)
scatter2d = ax2.scatter(X_embedded[:, 0], X_embedded[:, 1], c=t, s=5)
ax2.set_title("LPP Embedding (2D)")
ax2.set_xlabel("LPP 1")
ax2.set_ylabel("LPP 2")

plt.tight_layout()
plt.show()
```

# LPP vs PCA

```python
def make_linear_manifold(n_samples=1500, noise_std=0.05, random_state=0):
    """
    Create a 2D linear manifold embedded in 10D:
    - Start with (u, v) in R^2
    - Apply a random linear map to R^10
    - Add small Gaussian noise
    """
    rng = np.random.default_rng(random_state)

    # Intrinsic 2D coordinates (for coloring)
    u = rng.uniform(-2, 2, size=n_samples)
    v = rng.uniform(-1, 1, size=n_samples)
    Z = np.stack([u, v], axis=1)  # shape (n_samples, 2)

    # Random linear map from R^2 -> R^10
    A = rng.normal(size=(2, 10))  # (2, 10)
    X_clean = Z @ A               # (n_samples, 10)

    # Add small noise orthogonal-ish
    noise = noise_std * rng.normal(size=X_clean.shape)
    X = X_clean + noise

    # Use u (or norm of (u, v)) for coloring
    color = u  # you could also use np.linalg.norm(Z, axis=1)
    return X, Z, color
```

```python
# 1. Generate data
X, Z, color = make_linear_manifold(
    n_samples=1500,
    noise_std=0.05,
    random_state=42
)

print("Data shape:", X.shape)  # (n_samples, 10)
```

```python
# 2. PCA to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 3. LPP to 2D
lpp = LocalityPreservingProjections(
    n_components=2,
    n_neighbors=10,
    use_heat_kernel=True,
    t=1.0,
    reg=1e-3,
)
X_lpp = lpp.fit_transform(X)

print("PCA explained variance ratio:", pca.explained_variance_ratio_)
print("LPP eigenvalues:", lpp.eigenvalues_)
```

```python
# 4. Visual comparison
fig = plt.figure(figsize=(15, 4.5))

# (a) Original data: show first 3 coordinates as a 3D scatter
ax1 = fig.add_subplot(1, 3, 1, projection="3d")
sc1 = ax1.scatter(
    X[:, 0], X[:, 1], X[:, 2],
    c=color, s=5
)
ax1.set_title("Original Data (first 3 dims)")
ax1.set_xlabel("X0")
ax1.set_ylabel("X1")
ax1.set_zlabel("X2")

# (b) PCA embedding
ax2 = fig.add_subplot(1, 3, 2)
sc2 = ax2.scatter(
    X_pca[:, 0], X_pca[:, 1],
    c=color, s=5
)
ax2.set_title("PCA Embedding (2D)")
ax2.set_xlabel("PC1")
ax2.set_ylabel("PC2")

# (c) LPP embedding
ax3 = fig.add_subplot(1, 3, 3)
sc3 = ax3.scatter(
    X_lpp[:, 0], X_lpp[:, 1],
    c=color, s=5
)
ax3.set_title("LPP Embedding (2D)")
ax3.set_xlabel("LPP1")
ax3.set_ylabel("LPP2")

# Shared colorbar for all three
plt.tight_layout()
plt.show()
```
