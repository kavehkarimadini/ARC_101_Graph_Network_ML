This version follows the paper’s spirit:

* **DeepWalk as matrix factorization**: factorize $M \approx X^\top Y$, where $M = (A + A^2)/2$
* **Max-margin classifier**: multi-class hinge (Crammer–Singer style via “max competing class”)
* **Biased gradient**: instead of solving dual SVM for explicit $\alpha_{i}^{j}$, we use the **active constraint** $j^*$ from hinge loss (the violating class) to build the bias direction $(w_y - w_{j^*})$. This is effectively the **same geometry** as the paper’s bias term, but implemented in a stable SGD-friendly way.

---

## MMDW from scratch (PyTorch) on Cora

```python
import math
import torch
import torch.nn as nn
import torch.nn.functional as F

# PyG only for loading data
from torch_geometric.datasets import Planetoid
```

```python
# ----------------------------
# Utilities: build M = (A + A^2)/2
# ----------------------------
def build_transition_matrix(edge_index, num_nodes, device):
    """
    Build row-normalized adjacency (transition) matrix A as a dense tensor.
    Cora is small enough (2708 nodes) that dense is fine.
    """
    # Add self-loops for stability (common in graph methods)
    row, col = edge_index
    self_loops = torch.arange(num_nodes, device=device)
    row = torch.cat([row, self_loops], dim=0)
    col = torch.cat([col, self_loops], dim=0)

    A = torch.zeros((num_nodes, num_nodes), device=device, dtype=torch.float32)
    A[row, col] = 1.0

    # Row-normalize: transition probabilities
    deg = A.sum(dim=1, keepdim=True).clamp_min(1.0)
    A = A / deg
    return A


def build_deepwalk_matrix(A):
    """
    Paper uses a simplified DeepWalk matrix:
      M = (A + A^2)/2
    """
    A2 = A @ A
    M = 0.5 * (A + A2)
    return M
```

```python
# ----------------------------
# MMDW Model: embeddings + context + linear classifier
# ----------------------------
class MMDW(nn.Module):
    def __init__(self, num_nodes, emb_dim, num_classes):
        super().__init__()
        # X: node embeddings, Y: context embeddings (both k x |V| in paper; here |V| x k)
        self.X = nn.Embedding(num_nodes, emb_dim)
        self.Y = nn.Embedding(num_nodes, emb_dim)

        # W: linear multi-class weights (m x k)
        self.W = nn.Linear(emb_dim, num_classes, bias=False)

        # Init
        nn.init.normal_(self.X.weight, mean=0.0, std=0.1)
        nn.init.normal_(self.Y.weight, mean=0.0, std=0.1)
        nn.init.normal_(self.W.weight, mean=0.0, std=0.1)

    def node_emb(self):
        return self.X.weight  # [N, k]

    def ctx_emb(self):
        return self.Y.weight  # [N, k]

    def logits(self, x):
        return self.W(x)  # [B, C]
```

```python
# ----------------------------
# Losses
# ----------------------------
def matrix_factorization_loss(M, X, Y, lam=1e-4):
    """
    Frobenius loss: ||M - X Y^T||_F^2 + (lam/2)(||X||^2 + ||Y||^2)
    Note: X,Y are [N,k], so X Y^T is [N,N].
    """
    recon = X @ Y.t()
    diff = M - recon
    loss = (diff * diff).mean()
    reg = 0.5 * lam * (X.pow(2).mean() + Y.pow(2).mean())
    return loss + reg


def multiclass_hinge_loss(logits, y, C=1.0):
    """
    Crammer–Singer style hinge using the max competing class.
    For each i:
      loss_i = max(0, 1 + max_{j != y}(s_j) - s_y)
    """
    B, num_classes = logits.shape
    s_y = logits[torch.arange(B, device=logits.device), y]  # [B]

    # mask out correct class for max competing score
    mask = F.one_hot(y, num_classes=num_classes).bool()
    s_comp = logits.masked_fill(mask, float("-inf")).max(dim=1).values  # [B]

    margins = 1.0 + s_comp - s_y
    loss = torch.clamp(margins, min=0.0).mean()
    return C * loss, margins, s_comp


@torch.no_grad()
def compute_active_competing_class(logits, y):
    """
    For each sample, return j* = argmax_{j != y} s_j
    """
    B, C = logits.shape
    mask = F.one_hot(y, num_classes=C).bool()
    masked = logits.masked_fill(mask, float("-inf"))
    j_star = masked.argmax(dim=1)
    return j_star


def biased_gradient_step(model, train_idx, y_train, eta=1e-2, use_violation_scale=True):
    """
    Implements the paper's idea: for *violating* / support-like points, push x_i
    in direction (w_y - w_j*), where j* is the strongest competing class.

    We implement this as an *extra gradient-like nudging* step on X embeddings.
    This is a practical surrogate for the paper's alpha-weighted bias.
    """
    X = model.node_emb()  # [N,k]
    W = model.W.weight    # [C,k]

    x_i = X[train_idx]  # [T,k]
    logits = model.logits(x_i)  # [T,C]

    # find competing class j*
    j_star = compute_active_competing_class(logits, y_train)
    w_y = W[y_train]      # [T,k]
    w_j = W[j_star]       # [T,k]

    # margin violation: 1 + s_j* - s_y
    s_y = logits[torch.arange(len(train_idx), device=logits.device), y_train]
    s_j = logits[torch.arange(len(train_idx), device=logits.device), j_star]
    violation = 1.0 + s_j - s_y  # [T]

    active = (violation > 0).float().unsqueeze(1)  # [T,1]

    direction = (w_y - w_j)  # [T,k]
    if use_violation_scale:
        scale = violation.clamp_min(0.0).unsqueeze(1)  # [T,1]
        direction = direction * scale

    # Normalize direction for stability
    direction = direction / (direction.norm(dim=1, keepdim=True).clamp_min(1e-8))

    # Apply a small nudging step to embeddings (in-place on X.weight)
    # This is equivalent to adding the bias term to the gradient update.
    model.X.weight.data[train_idx] += eta * active * direction
```

```python
# ----------------------------
# Training / Evaluation
# ----------------------------
@torch.no_grad()
def evaluate(model, data, split="test"):
    model.eval()
    X = model.node_emb()
    logits = model.logits(X)
    pred = logits.argmax(dim=1)

    if split == "train":
        mask = data.train_mask
    elif split == "val":
        mask = data.val_mask
    else:
        mask = data.test_mask

    correct = (pred[mask] == data.y[mask]).float().mean().item()
    return correct
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load Cora (PyG only for data)
dataset = Planetoid(root="/tmp/Planetoid", name="Cora")
data = dataset[0].to(device)

N = data.num_nodes
C = dataset.num_classes

# Build M = (A + A^2)/2
A = build_transition_matrix(data.edge_index, N, device=device)
M = build_deepwalk_matrix(A)

# Model
emb_dim = 128
model = MMDW(num_nodes=N, emb_dim=emb_dim, num_classes=C).to(device)

# Hyperparams
lam_mf = 1e-3     # MF regularization
lr_mf = 5e-2      # MF optimizer LR
lr_svm = 1e-1     # SVM optimizer LR
C_svm = 1.0       # hinge scale
eta_bias = 5e-3   # biased gradient strength
outer_iters = 10  # alternate steps
svm_steps = 50    # optimize W per outer iter
mf_steps = 50     # optimize X,Y per outer iter

# Separate optimizers for clarity
opt_svm = torch.optim.SGD(model.W.parameters(), lr=lr_svm, momentum=0.9)
opt_mf = torch.optim.Adam([model.X.weight, model.Y.weight], lr=lr_mf)

train_idx = data.train_mask.nonzero(as_tuple=False).view(-1)
y_train = data.y[train_idx]

print(f"N={N}, classes={C}, train={int(data.train_mask.sum())}, "
      f"val={int(data.val_mask.sum())}, test={int(data.test_mask.sum())}")

for it in range(1, outer_iters + 1):
    # ---- (1) Fix X,Y -> train SVM W (max-margin) ----
    model.train()
    for _ in range(svm_steps):
        opt_svm.zero_grad()

        X = model.node_emb().detach()  # freeze embeddings for SVM step
        logits = model.logits(X[train_idx])

        hinge, _, _ = multiclass_hinge_loss(logits, y_train, C=C_svm)
        reg = 0.5 * (model.W.weight.pow(2).mean())
        loss_svm = hinge + reg

        loss_svm.backward()
        opt_svm.step()

    # ---- (2) Fix W -> update X,Y to factorize M, plus biased gradient effect ----
    model.train()
    for _ in range(mf_steps):
        opt_mf.zero_grad()

        # MF objective
        X = model.node_emb()
        Y = model.ctx_emb()
        loss_mf = matrix_factorization_loss(M, X, Y, lam=lam_mf)

        loss_mf.backward()
        opt_mf.step()

        # Apply biased gradient nudging on embeddings (support-like / violating points)
        biased_gradient_step(model, train_idx, y_train, eta=eta_bias, use_violation_scale=True)

    # ---- Evaluate ----
    tr = evaluate(model, data, "train")
    va = evaluate(model, data, "val")
    te = evaluate(model, data, "test")
    print(f"[Iter {it:02d}] train={tr:.4f} val={va:.4f} test={te:.4f}")

print("Done.")
```

---

## Notes / knobs you’ll likely tune

* `eta_bias`: stronger pushes embeddings to be more separable (too large can harm structure).
* `emb_dim`: 64/128/256 are common.
* `outer_iters`: paper often converges quickly; you can try 5–10.
* `mf_steps`, `svm_steps`: trade-off between structure-fitting and margin-fitting.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
try:
    import umap
    HAS_UMAP = True
except Exception:
    HAS_UMAP = False
```

```python
@torch.no_grad()
def get_embeddings_and_labels(model, data, split="all"):
    model.eval()
    X = model.node_emb().detach().cpu().numpy()
    y = data.y.detach().cpu().numpy()

    if split == "train":
        mask = data.train_mask.detach().cpu().numpy().astype(bool)
    elif split == "val":
        mask = data.val_mask.detach().cpu().numpy().astype(bool)
    elif split == "test":
        mask = data.test_mask.detach().cpu().numpy().astype(bool)
    else:
        mask = np.ones(len(y), dtype=bool)

    return X[mask], y[mask]


def _scatter_2d(Z, y, title):
    plt.figure(figsize=(7, 6))
    sc = plt.scatter(Z[:, 0], Z[:, 1], c=y, s=10, alpha=0.85)
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    cb = plt.colorbar(sc)
    cb.set_label("Class label")
    plt.tight_layout()
    plt.show()


def plot_tsne_embeddings(X, y, perplexity=30, n_iter=1000, seed=42, title="t-SNE"):
    tsne = TSNE(
        n_components=2,
        perplexity=perplexity,
        max_iter=n_iter,
        init="pca",
        learning_rate="auto",
        random_state=seed,
    )
    Z = tsne.fit_transform(X)
    _scatter_2d(Z, y, title)


def plot_umap_embeddings(X, y, n_neighbors=15, min_dist=0.1, seed=42, title="UMAP"):
    if not HAS_UMAP:
        raise ImportError("UMAP not found. Install with: pip install umap-learn")

    reducer = umap.UMAP(
        n_components=2,
        n_neighbors=n_neighbors,
        min_dist=min_dist,
        random_state=seed,
    )
    Z = reducer.fit_transform(X)
    _scatter_2d(Z, y, title)
```

```python
# Visualize embeddings
X_all, y_all = get_embeddings_and_labels(model, data, split="all")
plot_tsne_embeddings(X_all, y_all, perplexity=30, n_iter=1000, title="MMDW Embeddings (t-SNE, all nodes)")
```
