# **Practical HOPE Implementation**

*(Implementation aligned with the method described in the GSVD-based HOPE)*

We’ll implement the **full HOPE pipeline** for a chosen proximity measure (Katz).
This will follow the exact structure:

1. Build adjacency matrix A
2. Build $ M_g = I - \beta A $
3. Build $ M_l = \beta A $
4. Compute the **top-K generalized singular value decomposition** (GSVD)
5. Construct HOPE embeddings:

$$
U_s = [ \sqrt{\sigma_i} v_i^s ]
\quad\text{and}\quad
U_t = [ \sqrt{\sigma_i} v_i^t ]
$$

This implementation:

* works on any NetworkX directed graph
* constructs adjacency matrix A
* builds ( M_g ) and ( M_l ) as in the paper
* uses scipy’s sparse routines for efficiency
* uses truncated SVD on the transformed problem
* returns HOPE source & target embeddings

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from scipy.sparse import csc_matrix, identity
from scipy.sparse.linalg import eigs
from sklearn.manifold import TSNE
```

```python
# =========================================================
# 1. Load Cora (content + cites) into a directed NX graph
# =========================================================

def load_cora_graph(content_path="cora.content", cites_path="cora.cites"):
    """
    Load the Cora citation network.
    Assumes:
      - content: paper_id <features...> class_label
      - cites:   cited_id  citing_id   (edge direction: citing -> cited)
    """
    # Parse content file
    paper_ids = []
    labels_raw = []
    with open(content_path, "r") as f:
        for line in f:
            parts = line.strip().split()
            if not parts:
                continue
            paper_id = parts[0]
            label = parts[-1]
            paper_ids.append(paper_id)
            labels_raw.append(label)

    paper_ids = np.array(paper_ids)
    labels_raw = np.array(labels_raw)

    # Map labels to integers 0..C-1
    unique_labels = sorted(set(labels_raw))
    label_to_idx = {lab: i for i, lab in enumerate(unique_labels)}
    labels = np.array([label_to_idx[lab] for lab in labels_raw], dtype=int)

    # Build directed graph
    G = nx.DiGraph()
    for pid, lab in zip(paper_ids, labels):
        G.add_node(pid, label=lab)

    # Parse cites file
    with open(cites_path, "r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 2:
                continue
            cited, citing = parts
            # Use direction: citing -> cited
            if citing in G and cited in G:
                G.add_edge(citing, cited)

    return G, paper_ids, labels
```

```python
# =========================================================
# 2. HOPE (Katz) embedding implementation
# =========================================================

def hope_katz_embedding(G, beta=0.01, K=64):
    """
    HOPE embedding using the Katz proximity:
        S = (I - beta*A)^(-1) * beta*A

    G : directed NetworkX graph
    beta : Katz decay parameter (should be < 1 / spectral_radius(A))
    K : embedding dimension
    """
    nodes = list(G.nodes())
    idx = {nodes[i]: i for i in range(len(nodes))}
    N = len(nodes)

    # Adjacency matrix A (N x N)
    rows, cols, data = [], [], []
    for u, v in G.edges():
        rows.append(idx[u])
        cols.append(idx[v])
        data.append(1.0)
    A = csc_matrix((data, (rows, cols)), shape=(N, N))

    I = identity(N, format="csc")
    Mg = I - beta * A      # global part
    Ml = beta * A          # local part

    # Generalized eigenproblem:
    #   Ml Ml^T x = lambda Mg Mg^T x
    #   sigma = sqrt(lambda) are singular values of S
    M1 = (Ml @ Ml.T).tocsc()
    M2 = (Mg @ Mg.T).tocsc()

    # Top-K generalized eigenpairs
    vals, vecs = eigs(M1, k=K, M=M2)
    vals = np.real(vals)
    vs = np.real(vecs)

    # Numerical safety
    vals[vals < 0] = 0.0
    sigmas = np.sqrt(vals)

    # Compute right singular vectors (approx)
    vt = Ml.T @ vs
    for i in range(K):
        if sigmas[i] > 1e-12:
            vt[:, i] /= sigmas[i]
        else:
            vt[:, i] = 0.0

    # Build source & target embeddings
    Us = vs * np.sqrt(sigmas)
    Ut = vt * np.sqrt(sigmas)

    return Us, Ut, nodes
```

```python
# =========================================================
# 3. t-SNE visualization
# =========================================================

def visualize_tsne(Us, Ut=None, labels=None, title="HOPE (Katz) + t-SNE on Cora"):
    """
    t-SNE visualization for HOPE embeddings.

    Us: source embedding (N x K)
    Ut: target embedding (optional, N x K)
    labels: optional labels for coloring (length N)
    """
    if Ut is not None:
        X = np.concatenate([Us, Ut], axis=1)
    else:
        X = Us

    tsne = TSNE(
        n_components=2,
        perplexity=30,
        learning_rate=200,
        init="random",
        random_state=42
    )
    X_2d = tsne.fit_transform(X)

    plt.figure(figsize=(8, 6))
    if labels is None:
        plt.scatter(X_2d[:, 0], X_2d[:, 1], s=8, alpha=0.8)
    else:
        scatter = plt.scatter(
            X_2d[:, 0],
            X_2d[:, 1],
            c=labels,
            s=8,
            alpha=0.8,
            cmap="tab10"
        )
        plt.colorbar(scatter, label="Class")

    plt.title(title)
    plt.xlabel("t-SNE dim 1")
    plt.ylabel("t-SNE dim 2")
    plt.tight_layout()
    plt.show()
```

```python
# Adjust paths if needed
G, paper_ids, labels = load_cora_graph("./Data/cora/cora.content", "./Data/cora/cora.cites")
print(f"Loaded Cora: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")
```

```python
# Compute HOPE embeddings
Us, Ut, nodes = hope_katz_embedding(G, beta=0.01, K=64)
```

```python
# Reorder labels to match `nodes` order
# (nodes is the list of node IDs in the same order as rows in Us/Ut)
label_dict = {node: G.nodes[node]["label"] for node in G.nodes()}
labels_ordered = np.array([label_dict[n] for n in nodes])
```

```python
# Visualize concatenated source + target embeddings
visualize_tsne(Us, Ut, labels=labels_ordered,
               title="HOPE (Katz) + t-SNE on Cora")
```
