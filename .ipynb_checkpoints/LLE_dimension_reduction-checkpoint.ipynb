```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # necessary for 3D projection
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding
```

# **Solving for the Reconstruction Weights ( W )**

> The optimal weights $ W_{ij} $ subject to these constraints are found by **solving a least-squares problem**.

This is the core computational step in LLE.

---

## **The Least-Squares Problem**

For each point $ X_i $, we solve:

$
\min_{W_{ij}} \left| X_i - \sum_j W_{ij} X_j \right|^2
$

with:

1. $ W_{ij} = 0 $ if $ X_j $ is not a neighbor
2. $ \sum_{j\in \mathcal{N}(i)} W_{ij} = 1 $

This is done **separately** for each data point.

### **Key geometric interpretation**

Let $ Z_j = X_j - X_i $.
Then:

$
X_i - \sum_j W_{ij} X_j
= X_i - \sum_j W_{ij}(X_i + Z_j)
= - \sum_j W_{ij} Z_j
$

The reconstruction depends only on the **relative neighbor offsets**.

Thus the local covariance matrix:

$
C_{jk} = (X_j - X_i)^\top (X_k - X_i)
$

is used to solve:

$
\min_{W} W^\top C W
$

subject to $ \sum_j W_j = 1 $.

This is a constrained quadratic optimization that has a closed-form solution.

---

# **In short:**

* LLE expresses local neighborhoods in a local coordinate system relative to each point.
* It finds weights that reconstruct each point from its neighbors.
* This is solved using a small $K \times K$ linear system for each point.
* These weights encode intrinsic local geometry of the manifold.

---


# **Global Embedding via Eigenvectors**

This is the second major optimization in LLE.

> LLE constructs a neighborhood-preserving mapping … by choosing
> low-dimensional coordinates $Y_i$ to minimize
> $
> \Phi(Y) = \sum_i \left| Y_i - \sum_j W_{ij} Y_j \right|^2
> $

### **Interpretation**

Now that the weights (W) encode the local geometry, LLE finds low-dimensional points $Y_i$ such that:

* Each point in low-D is reconstructed from neighbors **using the same weights**.
* This forces the low-D embedding to preserve the local geometric structure.

### **Why eigenvectors?**

Expanding the embedding cost:

$
\Phi(Y) = \sum_i \left| Y_i - \sum_j W_{ij} Y_j \right|^2
$

leads to a quadratic form:

$
\Phi(Y) = Y^\top M Y,
$

where

$
M = (I - W)^\top (I - W)
$

This is a symmetric, positive semi-definite matrix.

The solution:

* The bottom **d+1** eigenvectors of (M)
* Discard the first (eigenvalue 0) because it corresponds to a constant shift
* The next **d** eigenvectors are the embedding coordinates

Thus:
**LLE embedding = bottom nonzero eigenvectors of a sparse matrix.**

---

# Why does this work?

Because the eigenvectors give the smoothest functions over the graph defined by ( W ), preserving local relations defined earlier.

---

# **Important properties**

* No need for pairwise geodesic distances (unlike Isomap).
* The matrix ( M ) is very sparse → computationally efficient.
* Only a single sparse eigen-decomposition is required.

---


```python
class LLE:
    def __init__(self, n_neighbors=10, n_components=2, reg=1e-3):
        """
        n_neighbors : int
            Number of nearest neighbours (K)
        n_components : int
            Output embedding dimension (d)
        reg : float
            Regularization strength for local covariance matrices
        """
        self.n_neighbors = n_neighbors
        self.n_components = n_components
        self.reg = reg
        self.embedding_ = None
        self.neighbors_ = None
        self.weights_ = None

    def _compute_neighbors(self, X):
        """
        Brute-force KNN using Euclidean distance.
        Returns indices of shape (N, K)
        """
        N = X.shape[0]
        # Compute full distance matrix (N x N)
        # ||x-y||^2 = ||x||^2 + ||y||^2 - 2 x·y
        sq_norms = np.sum(X**2, axis=1, keepdims=True)  # (N,1)
        dist2 = sq_norms + sq_norms.T - 2 * X @ X.T     # (N,N)
        
        # argsort rows, skip index 0 (self), take next K
        neighbors = np.argsort(dist2, axis=1)[:, 1:self.n_neighbors+1]
        return neighbors

    def _compute_weights(self, X, neighbors):
        """
        Compute reconstruction weights W (N x N) given X and neighbor indices.
        Each row i has nonzeros only at columns neighbors[i].
        """
        N, D = X.shape
        K = neighbors.shape[1]
        W = np.zeros((N, N), dtype=np.float64)

        for i in range(N):
            nbr_idx = neighbors[i]               # shape (K,)
            Xi = X[i]                            # shape (D,)
            Ni = X[nbr_idx]                      # shape (K, D)

            # Shift neighbors to local coordinates
            Z = Ni - Xi                           # (K, D)

            # Local covariance C = Z Z^T (K x K)
            C = Z @ Z.T

            # Regularization (to handle rank deficiency / noise)
            trace = np.trace(C)
            if trace > 0:
                C += self.reg * trace * np.eye(K)
            else:
                C += self.reg * np.eye(K)

            # Solve C w = 1 with sum(w) = 1 constraint
            # Closed form: w = C^{-1} 1 / (1^T C^{-1} 1)
            ones = np.ones(K, dtype=np.float64)
            w = np.linalg.solve(C, ones)
            w /= w.sum()

            # Store in big weight matrix
            W[i, nbr_idx] = w

        return W

    def _compute_embedding(self, W):
        """
        Given weight matrix W (N x N), compute embedding Y (N x d)
        from the bottom non-trivial eigenvectors of M = (I - W)^T (I - W).
        """
        N = W.shape[0]
        I = np.eye(N)
        M = (I - W)
        M = M.T @ M   # symmetric, positive semi-definite

        # Compute eigenvalues/vectors of M
        # eigh returns them in ascending order of eigenvalues
        evals, evecs = np.linalg.eigh(M)

        # Skip the first eigenvector (eigenvalue 0, constant)
        # Take next n_components as embedding
        idx = np.argsort(evals)
        evecs = evecs[:, idx]
        embedding = evecs[:, 1:self.n_components+1]

        return embedding

    def fit_transform(self, X):
        """
        Run full LLE and return low-dimensional embedding.
        """
        X = np.asarray(X, dtype=np.float64)
        self.neighbors_ = self._compute_neighbors(X)
        self.weights_ = self._compute_weights(X, self.neighbors_)
        self.embedding_ = self._compute_embedding(self.weights_)
        return self.embedding_
```

```python
# 1. Generate Swiss roll data
n_samples = 1500
X, t = make_swiss_roll(n_samples=n_samples, noise=0.05)
# t is the intrinsic "unrolled" coordinate -> great for coloring
```

```python
# 2. Run LLE from scratch
lle = LLE(n_neighbors=12, n_components=2, reg=1e-3)
Y = lle.fit_transform(X)
```

```python
# 3. Plot results

# --- 3D Swiss roll ---
fig = plt.figure(figsize=(12, 5))

ax = fig.add_subplot(1, 2, 1, projection="3d")
p = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, s=5)
ax.set_title("Swiss Roll (3D)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(p, ax=ax, label="Intrinsic parameter t")

# --- 2D LLE embedding ---
ax2 = fig.add_subplot(1, 2, 2)
sc = ax2.scatter(Y[:, 0], Y[:, 1], c=t, s=5)
ax2.set_title("LLE Embedding (2D)")
ax2.set_xlabel("Component 1")
ax2.set_ylabel("Component 2")
fig.colorbar(sc, ax=ax2, label="Intrinsic parameter t")

plt.tight_layout()
plt.show()
```

- Left plot (3D): Curled Swiss roll in 3D, colored by intrinsic parameter t.

- Right plot (2D): LLE’s 2D embedding, where the roll is “unwrapped” and the color gradient should stay smooth → meaning LLE preserved the manifold structure.

# Using sklearn

```python
# 2. Run LLE using sklearn
n_neighbors = 12
n_components = 2

lle = LocallyLinearEmbedding(
    n_neighbors=n_neighbors,
    n_components=n_components,
    method="standard",   # plain LLE
    eigen_solver="auto",
    random_state=0
)
Y = lle.fit_transform(X)   # shape: (n_samples, 2)

print("Embedding shape:", Y.shape)
```

```python
# 3. Plot results

fig = plt.figure(figsize=(12, 5))

# --- 3D Swiss roll ---
ax = fig.add_subplot(1, 2, 1, projection="3d")
p = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, s=5)
ax.set_title(f"Swiss Roll (3D)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(p, ax=ax, label="Intrinsic parameter t")

# --- 2D LLE embedding ---
ax2 = fig.add_subplot(1, 2, 2)
sc = ax2.scatter(Y[:, 0], Y[:, 1], c=t, s=5)
ax2.set_title(f"LLE Embedding (2D) – sklearn (K={n_neighbors})")
ax2.set_xlabel("Component 1")
ax2.set_ylabel("Component 2")
fig.colorbar(sc, ax=ax2, label="Intrinsic parameter t")

plt.tight_layout()
plt.show()
```
