## 1. Setup and Data: Karate Club → Signed Network

We’ll:

* Load KarateClub from PyG
* Use the community labels `data.y` to generate **signs**:

  * +1 if two nodes are in the **same community**
  * −1 if they are in **different communities**

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # needed for 3D projection
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.datasets import KarateClub
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 1) Load Karate Club graph from PyG (only for data, not for models)
dataset = KarateClub()
data = dataset[0]

num_nodes = data.num_nodes
edge_index = data.edge_index  # [2, E]
labels = data.y               # community labels per node

print(f"#nodes = {num_nodes}, #edges = {edge_index.size(1)}")
```

Now build a **signed adjacency matrix** $A \in {-1, 0, +1}^{n \times n}$:


```python
# 2) Build signed adjacency matrix A: +1 if same community, -1 otherwise
A = torch.zeros((num_nodes, num_nodes), dtype=torch.float32)

# Make the graph undirected explicitly
src, dst = edge_index
for i, j in zip(src.tolist(), dst.tolist()):
    if labels[i] == labels[j]:
        sign = 1.0
    else:
        sign = -1.0
    A[i, j] = sign
    A[j, i] = sign  # undirected

# Remove self-loops just to be clean
for i in range(num_nodes):
    A[i, i] = 0.0

A = A.to(device)
```

Now construct **A⁺, A⁻, degrees and Laplacians**:

```python
# A+ and A- (absolute positive / negative adjacency)
A_pos = torch.clamp(A, min=0.0)
A_neg = -torch.clamp(A, max=0.0)

# Degree vectors
deg_pos = A_pos.sum(dim=1)  # [n]
deg_neg = A_neg.sum(dim=1)  # [n]

# Diagonal degree matrices
D_pos = torch.diag(deg_pos)
D_neg = torch.diag(deg_neg)

# Laplacians
L_pos = D_pos - A_pos
L_neg = D_neg - A_neg
```

---

## 2. Penalty Matrix (P)

As in the paper:

$$
P_{ij} =
\begin{cases}
1 & A_{ij} = 0 \\
\beta & A_{ij} > 0 \\
\gamma \cdot \beta & A_{ij} < 0
\end{cases}
$$

We’ll choose some reasonable hyperparams:

```python
beta = 25.0    # larger penalty for observed vs unobserved
gamma = 3.0    # extra penalty for negative vs positive

P = torch.ones_like(A)

P[A > 0] = beta
P[A < 0] = gamma * beta

P = P.to(device)
```

We also need **L = L⁺ − γ L⁻** for the pairwise constraint:

```python
L = L_pos - gamma * L_neg
L = L.to(device)
```

---

## 3. DNE-SBP Model: Stacked Autoencoder

We’ll implement a fully-connected SAE with tanh activations, as in the paper:

```python
class DNESBP(nn.Module):
    def __init__(self, num_nodes, hidden_dims):
        """
        num_nodes: input dimension (n)
        hidden_dims: list, e.g. [256, 64]
        """
        super().__init__()
        dims = [num_nodes] + hidden_dims
        # Encoder
        self.encoder = nn.ModuleList()
        for in_dim, out_dim in zip(dims[:-1], dims[1:]):
            self.encoder.append(nn.Linear(in_dim, out_dim))
        # Decoder (mirror)
        dec_dims = list(reversed(dims))
        self.decoder = nn.ModuleList()
        for in_dim, out_dim in zip(dec_dims[:-1], dec_dims[1:]):
            self.decoder.append(nn.Linear(in_dim, out_dim))

    def forward(self, X):
        """
        X: [n, n] adjacency matrix
        Returns:
            X_hat: reconstruction of X
            H: deepest hidden representation [n, d]
        """
        h = X
        for layer in self.encoder:
            h = torch.tanh(layer(h))
        H = h  # deepest hidden
        for layer in self.decoder:
            h = torch.tanh(layer(h))
        X_hat = h
        return X_hat, H
```

Instantiate the model:

```python
hidden_dims = [256, 64]  # 2-layer SAE like in the paper (7118-256-64)
model = DNESBP(num_nodes=num_nodes, hidden_dims=hidden_dims).to(device)
```

---

## 4. Loss Function: Reconstruction + Structural Balance + Regularization

We implement:

* **Reconstruction term** (first-layer-like)

  $$
  J_1 = \frac{1}{2n} | (\hat{A} - A) \odot P |_F^2
  $$

* **Pairwise constraint (structural balance)**

  $$
  J_{\text{pair}} = \frac{1}{n} \mathrm{Tr}(H^T L H)
  $$

* **L2 regularization** on weights

We use **α** and **λ** as hyperparameters:

```python
def dne_sbp_loss(model, A, A_hat, H, P, L, alpha=10.0, lam=1e-4):
    """
    model: DNESBP
    A: [n, n] signed adjacency
    A_hat: [n, n] reconstruction
    H: [n, d] deepest hidden
    P: [n, n] penalty matrix
    L: [n, n] structural-balance Laplacian
    alpha: weight for pairwise constraint
    lam: weight for L2 regularization
    """
    n = A.size(0)

    # 1) Reconstruction with penalty
    diff = (A_hat - A) * P
    J_rec = 0.5 / n * torch.sum(diff * diff)

    # 2) Pairwise constraint (structural balance)
    # Tr(H^T L H) / n
    J_pair = torch.trace(H.t() @ L @ H) / n

    # 3) L2 regularization
    J_reg = 0.0
    for p in model.parameters():
        if p.dim() >= 2:  # only weights, not biases
            J_reg = J_reg + torch.sum(p * p)
    J_reg = 0.5 * J_reg

    return J_rec + alpha * J_pair + lam * J_reg, (J_rec, J_pair, J_reg)
```

---

## 5. Training Loop (Unsupervised DNE-SBP)

We treat each **row of A** as an input sample (node’s adjacency pattern).
For simplicity, we **train on the whole matrix at once** (Karate is tiny).

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

A_input = A  # [n, n]

num_epochs = 1000
alpha = 10.0   # pairwise constraint weight
lam = 1e-4     # regularization weight

for epoch in range(1, num_epochs + 1):
    model.train()
    optimizer.zero_grad()

    A_hat, H = model(A_input)
    loss, (J_rec, J_pair, J_reg) = dne_sbp_loss(
        model, A, A_hat, H, P, L, alpha=alpha, lam=lam
    )

    loss.backward()
    optimizer.step()

    if epoch % 100 == 0 or epoch == 1:
        print(
            f"Epoch {epoch:04d} | "
            f"Loss={loss.item():.4f} | "
            f"Rec={J_rec.item():.4f} | "
            f"Pair={J_pair.item():.4f} | "
            f"Reg={J_reg.item():.4f}"
        )
```

```python
# After training, H is your node embedding matrix [n, d]
model.eval()
with torch.no_grad():
    _, H_final = model(A_input)
print("Final embedding shape:", H_final.shape)
```

---

## 6. Link Sign Prediction from Embeddings (Hadamard Edge Features)

Now we’ll show how to use the learned embeddings `H_final` for **sign prediction**:

1. Build a list of edges `(i, j)` and labels `y`:

   * y = 1 for positive
   * y = 0 for negative
2. Construct **Hadamard edge features**: $ h_i \odot h_j $
3. Train a simple **logistic regression** classifier in PyTorch.

```python
import random

# 1) Build edge list and labels from A
edges = []
labels_edge = []

for i in range(num_nodes):
    for j in range(i + 1, num_nodes):  # upper triangle only
        if A[i, j].item() != 0.0:
            edges.append((i, j))
            label = 1 if A[i, j].item() > 0 else 0
            labels_edge.append(label)

edges = torch.tensor(edges, dtype=torch.long, device=device)   # [E, 2]
labels_edge = torch.tensor(labels_edge, dtype=torch.float32, device=device)  # [E]

num_edges = edges.size(0)
print("Num signed edges:", num_edges)
```

```python
# 2) Build Hadamard features
with torch.no_grad():
    h_i = H_final[edges[:, 0]]  # [E, d]
    h_j = H_final[edges[:, 1]]  # [E, d]
    X_edge = h_i * h_j         # Hadamard product [E, d]

# 3) Train/test split
perm = torch.randperm(num_edges)
split = int(0.8 * num_edges)
train_idx = perm[:split]
test_idx = perm[split:]

X_train = X_edge[train_idx]
y_train = labels_edge[train_idx]
X_test = X_edge[test_idx]
y_test = labels_edge[test_idx]
```

Define a **logistic regression classifier**:

```python
class EdgeLogisticRegression(nn.Module):
    def __init__(self, in_dim):
        super().__init__()
        self.linear = nn.Linear(in_dim, 1)

    def forward(self, x):
        logits = self.linear(x)
        return logits.squeeze(-1)  # [E]
```

Train and evaluate (simple accuracy to keep code compact):

```python
clf = EdgeLogisticRegression(H_final.size(1)).to(device)
opt_clf = torch.optim.Adam(clf.parameters(), lr=0.01)
bce = nn.BCEWithLogitsLoss()

for epoch in range(1, 501):
    clf.train()
    opt_clf.zero_grad()

    logits = clf(X_train)
    loss_clf = bce(logits, y_train)
    loss_clf.backward()
    opt_clf.step()

    if epoch % 100 == 0 or epoch == 1:
        with torch.no_grad():
            clf.eval()
            test_logits = clf(X_test)
            preds = (torch.sigmoid(test_logits) > 0.5).float()
            acc = (preds == y_test).float().mean().item()
        print(f"[LR] Epoch {epoch:04d} | Loss={loss_clf.item():.4f} | Test Acc={acc:.4f}")
```

On this toy graph, you should see **very high accuracy**, since we *used the communities* themselves to define the signs.

For a real experiment, you’d:

* use a true signed dataset (Epinions, Slashdot, Wiki, etc.),
* build A directly from their +/− links,
* and evaluate using **AUC** and **AP** as in the paper.

## 7. How to Extend This to Full Paper-Fidelity DNE-SBP

If you want to match the paper *exactly*, here’s what to extend:

1. **Layer-wise losses**

   * Instead of only constraining the last hidden layer:

     * store all hidden layer outputs $H^{(k)}$
     * reconstruct *per layer* (you’d need partial decoders)
     * sum J₁⁽ᵏ⁾ + αₖ(J₂⁽ᵏ⁾ + γₖJ₃⁽ᵏ⁾) + λₖJ₄⁽ᵏ⁾

2. **Greedy layer-wise pre-training**

   * Train first AE (A → H¹ → Ā)
   * Freeze it, train second AE on H¹, etc.

3. **Hyperparameters per layer**

   * Distinct αₖ, λₖ, γₖ per layer, as in Table III of the paper.

The structure of the code stays the same; you just refactor the model to expose intermediate reconstructions and apply the same logic layer-by-layer.

Below is **pure PyTorch + scikit-learn + matplotlib** code to:

1. Take your learned embeddings `H_final`
2. Reduce them to **3D with t-SNE**
3. Plot them as a **3D scatter**, colored by clusters

You can choose:

* **Ground-truth clusters** (e.g. `labels` from KarateClub)
* **K-means clusters** on the embeddings

---

## ✅ 1. t-SNE (3D) + Ground-Truth Cluster Colors

This assumes you already have:

* `H_final`: tensor of shape `[num_nodes, d]` (from DNE-SBP)
* `labels`: tensor of shape `[num_nodes]` with integer cluster IDs

```python
# Assume these already exist:
# H_final: [num_nodes, d] from your trained DNE-SBP
# labels: [num_nodes] ground-truth or cluster assignments
H_np = H_final.detach().cpu().numpy()
y_np = labels.detach().cpu().numpy()

# 3D t-SNE
tsne = TSNE(
    n_components=3,
    init="random",
    learning_rate="auto",
    perplexity=10,     # tweak if needed (depends on num_nodes)
    random_state=42,
)
H_tsne_3d = tsne.fit_transform(H_np)  # shape: [num_nodes, 3]

# 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

scatter = ax.scatter(
    H_tsne_3d[:, 0],
    H_tsne_3d[:, 1],
    H_tsne_3d[:, 2],
    c=y_np,
    cmap="tab10",      # color by cluster id
    s=40,
    alpha=0.9,
)

ax.set_title("DNE-SBP Embeddings (3D t-SNE)")
ax.set_xlabel("t-SNE 1")
ax.set_ylabel("t-SNE 2")
ax.set_zlabel("t-SNE 3")

# Optional: legend for discrete cluster IDs
handles, _ = scatter.legend_elements()
ax.legend(handles, [f"Cluster {i}" for i in range(len(np.unique(y_np)))],
          title="Clusters", loc="best")

plt.tight_layout()
plt.show()
```

## ✅ 2. If You Want to Color by **K-Means Clusters on Embeddings**

If you’d rather color by clusters **discovered from the embedding itself**:


```python
from sklearn.cluster import KMeans

H_np = H_final.detach().cpu().numpy()

# Choose number of clusters (e.g., 2 for KarateClub)
k = 2
kmeans = KMeans(n_clusters=k, random_state=42)
cluster_ids = kmeans.fit_predict(H_np)   # [num_nodes]

# t-SNE as before
tsne = TSNE(
    n_components=3,
    init="random",
    learning_rate="auto",
    perplexity=10,
    random_state=42,
)
H_tsne_3d = tsne.fit_transform(H_np)

# Plot with cluster_ids as colors
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

scatter = ax.scatter(
    H_tsne_3d[:, 0],
    H_tsne_3d[:, 1],
    H_tsne_3d[:, 2],
    c=cluster_ids,
    cmap="tab10",
    s=40,
    alpha=0.9,
)

ax.set_title("DNE-SBP Embeddings (3D t-SNE, K-Means Colors)")
ax.set_xlabel("t-SNE 1")
ax.set_ylabel("t-SNE 2")
ax.set_zlabel("t-SNE 3")

handles, _ = scatter.legend_elements()
ax.legend(handles, [f"Cluster {i}" for i in range(k)],
          title="K-Means Cluster", loc="best")

plt.tight_layout()
plt.show()
```
