```python
import random
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Load Zachary's Karate Club graph from networkx
G = nx.karate_club_graph()
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
```

### Map nodes to indices for embedding matrices

```python
# Map nodes to indices [0, 1, ..., N-1]
nodes = list(G.nodes())
node_to_id = {node: i for i, node in enumerate(nodes)}
id_to_node = {i: node for node, i in node_to_id.items()}
num_nodes = len(nodes)
```

# Generate Random Walks
> - DeepWalk core: short, truncated random walks.

```python
def random_walk(G, start_node, walk_length):
    walk = [start_node]
    current = start_node
    for _ in range(walk_length - 1):
        neighbors = list(G.neighbors(current))
        if len(neighbors) == 0:
            break
        current = random.choice(neighbors)
        walk.append(current)
    return walk
```

### Generate multiple walks per node:

```python
def generate_walks(G, num_walks_per_node=10, walk_length=40):
    walks = []
    nodes = list(G.nodes())
    for _ in range(num_walks_per_node):
        random.shuffle(nodes)  # good for SGD-like behavior
        for node in nodes:
            walk = random_walk(G, node, walk_length)
            walks.append(walk)
    return walks
```

```python
# Example
walks = generate_walks(G, num_walks_per_node=10, walk_length=20)
print("Number of walks:", len(walks))
print("First walk:", walks[0])
```

- At this point you have a “corpus” of walks, where each walk is like a sentence of “words” (nodes).

# Build Skip-Gram Training Pairs

> - From each walk, create (center, context) pairs using a window of size window_size.

```python
def build_skipgram_pairs(walks, window_size=5):
    pairs = []
    for walk in walks:
        walk_ids = [node_to_id[n] for n in walk]
        for i, center in enumerate(walk_ids):
            # context window [i - w, i + w]
            for j in range(max(0, i - window_size), min(len(walk_ids), i + window_size + 1)):
                if i == j:
                    continue
                context = walk_ids[j]
                pairs.append((center, context))
    return pairs
```

```python
pairs = build_skipgram_pairs(walks, window_size=5)
print("Number of training pairs:", len(pairs))
print("Example pair (center_id, context_id):", pairs[0])
```

# Negative Sampling Distribution

> - We’ll approximate word2vec’s negative sampling: sample negatives according to degree^0.75.

```python
# Build frequency distribution (use node degree as proxy)
degrees = np.array([G.degree(id_to_node[i]) for i in range(num_nodes)], dtype=np.float64)
freq = degrees ** 0.75
freq = freq / freq.sum()

def sample_negative(num_samples):
    # Returns array of node ids sampled as negatives
    return np.random.choice(num_nodes, size=num_samples, p=freq)
```

# Initialize Embeddings

> We have two embedding matrices:

>> 1. `W_in`: input embeddings (center node)
>> 2. `W_out`: output embeddings (context node)

```python
embedding_dim = 64
rng = np.random.default_rng(42)

W_in = (rng.random((num_nodes, embedding_dim)) - 0.5) / embedding_dim
W_out = (rng.random((num_nodes, embedding_dim)) - 0.5) / embedding_dim
```

# Training: Skip-Gram with Negative Sampling (SGNS)

> Loss for one (center, positive) pair with k negatives:
$
L = -\log \sigma(u_p \cdot v_c) - \sum_{i=1}^k \log \sigma(-u_{n_i} \cdot v_c)
$

>We’ll do **plain SGD**, one pair at a time (Karate is tiny, so this is fine).

```python
def train_deepwalk(
    pairs,
    num_nodes,
    embedding_dim=64,
    num_epochs=5,
    neg_samples=5,
    learning_rate=0.025,
    print_every=10000
):
    rng = np.random.default_rng(42)
    # Initialize embeddings
    W_in = (rng.random((num_nodes, embedding_dim)) - 0.5) / embedding_dim
    W_out = (rng.random((num_nodes, embedding_dim)) - 0.5) / embedding_dim

    pair_count = len(pairs)

    for epoch in range(num_epochs):
        random.shuffle(pairs)
        total_loss = 0.0

        for step, (center, context) in enumerate(pairs):
            # Sample negatives
            negatives = sample_negative(neg_samples)

            v_c = W_in[center]                    # (d,)
            u_p = W_out[context]                  # (d,)
            u_ns = W_out[negatives]               # (k, d)

            # Positive score and sigma
            pos_score = np.dot(u_p, v_c)          # scalar
            pos_sig = 1.0 / (1.0 + np.exp(-pos_score))

            # Negative scores and sigmas
            neg_scores = u_ns @ v_c               # (k,)
            neg_sig = 1.0 / (1.0 + np.exp(-neg_scores))

            # Loss = -log σ(pos) - Σ log σ(-neg)
            # log σ(-x) = log(1 - σ(x))
            loss = -np.log(pos_sig + 1e-10) - np.sum(np.log(1.0 - neg_sig + 1e-10))
            total_loss += loss

            # Gradients:
            # dL/dv_c = (σ(pos) - 1)*u_p + Σ σ(neg)*u_n
            grad_v = (pos_sig - 1.0) * u_p + (neg_sig @ u_ns)

            # dL/du_p = (σ(pos) - 1)*v_c
            grad_u_p = (pos_sig - 1.0) * v_c

            # dL/du_n = σ(neg_i)*v_c
            grad_u_ns = (neg_sig[:, np.newaxis] * v_c[np.newaxis, :])  # (k, d)

            # SGD updates
            W_in[center] -= learning_rate * grad_v
            W_out[context] -= learning_rate * grad_u_p
            W_out[negatives] -= learning_rate * grad_u_ns

            if (step + 1) % print_every == 0:
                avg_loss = total_loss / print_every
                print(f"Epoch {epoch+1}, step {step+1}/{pair_count}, avg loss = {avg_loss:.4f}")
                total_loss = 0.0

        # End of epoch
        avg_epoch_loss = total_loss / max(1, (pair_count % print_every))
        print(f"Epoch {epoch+1} finished, remaining avg loss = {avg_epoch_loss:.4f}")

    return W_in, W_out
```

```python
# Train on Karate
W_in, W_out = train_deepwalk(
    pairs,
    num_nodes=num_nodes,
    embedding_dim=64,
    num_epochs=5,
    neg_samples=5,
    learning_rate=0.025,
    print_every=5000
)
```

Here:

* `W_in` is usually what we use as the **final node embeddings**.
* You can also use `W_in + W_out` or other combos, but `W_in` alone is common.

# Get Node Embeddings and Visualize

>Extract an embedding per node:

```python
node_embeddings = {id_to_node[i]: W_in[i] for i in range(num_nodes)}
```

### Let’s do a simple 2D visualization with PCA to see communities:

```python
# nodes, G, and W_in assumed from previous code
X = np.vstack([W_in[node_to_id[n]] for n in nodes])

# 2D projection with PCA
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# Get the ground-truth club labels from NetworkX
club_attr = nx.get_node_attributes(G, "club")  # node -> "Mr. Hi" or "Officer"

# Map club names to colors
club_to_color = {"Mr. Hi": 0, "Officer": 1}
colors = [club_to_color[club_attr[n]] for n in nodes]

plt.figure(figsize=(6, 6))

# Scatter with colors according to club membership
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=colors, cmap="coolwarm")

# Add node ids as text
for i, node in enumerate(nodes):
    x, y = X_2d[i, 0], X_2d[i, 1]
    plt.text(x + 0.01, y + 0.01, str(node), fontsize=9)

# Legend
handles = [
    plt.Line2D([0], [0], marker='o', linestyle='', label='Mr. Hi',  color=scatter.cmap(scatter.norm(0))),
    plt.Line2D([0], [0], marker='o', linestyle='', label='Officer', color=scatter.cmap(scatter.norm(1))),
]
plt.legend(handles=handles, title="Club")

plt.title("DeepWalk Embeddings of Karate Club (colored by club)")
plt.axis('equal')
plt.show()
```

You should see nodes form **clusters** roughly matching the classic split in the Karate Club.

---

* **Random walks** → `generate_walks`
* **Treat walks like sentences** → `build_skipgram_pairs`
* **Skip-Gram objective** → `train_deepwalk` (center → context prediction)
* **Fast approximation** → we used **negative sampling** instead of hierarchical softmax (the paper uses hierarchical softmax, but conceptually it’s the same “make co-occurring nodes probable”).

# stellargraph ( supports DeepWalk, Node2Vec, GCNs, etc.)

```python
from stellargraph.data import BiasedRandomWalk
from stellargraph import StellarGraph
from gensim.models import Word2Vec
```

```python
# Create graph
G_nx = nx.karate_club_graph()
G = StellarGraph.from_networkx(G_nx)
```

```python
# Random walks
rw = BiasedRandomWalk(G)
walks = rw.run(
    nodes=G.nodes(),
    n=10,             # walks per node
    length=40         # walk length
)
```

```python
# Train Skip-Gram (DeepWalk)
model = Word2Vec(
    walks,
    vector_size=64,
    window=5,
    sg=1,             # skip-gram
    workers=4,
    min_count=0
)
```

```python
# Get embeddings
emb = []
for i in range(len(G_nx.nodes)):
    emb.append(model.wv[i])  # stellargraph uses string node IDs

emb = np.array(emb)
```

```python
# PCA visualization
pca = PCA(n_components=2)
X_2d = pca.fit_transform(emb)

club_attr = nx.get_node_attributes(G_nx, "club")
club_to_color = {"Mr. Hi": 0, "Officer": 1}
colors = [club_to_color[club_attr[i]] for i in G_nx.nodes()]

plt.figure(figsize=(6, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=colors, cmap="coolwarm")
plt.title("DeepWalk via StellarGraph")
plt.show()
```
